Create tuple from Buildings in a single tract where the tract has no Roads or Non-Buildings

```python
def make_row_NO_Roads(tract):
    row={}
    row['Aeroway']=len(tract[tract.aeroway.notnull()])
    row['Amenity-Substance']=len(tract[tract.amenity.isin(Amen_Substance)])
    row['Amenity-Education']=len(tract[tract.amenity.isin(Amen_Education)])
    row['Amenity-Transportation']=len(tract[tract.amenity.isin(Amen_Trans)])
    row['Amenity-Financial']=len(tract[tract.amenity.isin(Amen_Fin)])
    row['Amenity-Healthcare']=len(tract[tract.amenity.isin(Amen_HC)])
    row['Amenity-Entertainment, Arts, Culture']=len(tract[tract.amenity.isin(Amen_EAC)])
    row['Amenity-Public Service']=len(tract[tract.amenity.isin(Amen_PS)])
    row['Amenity-Facility']=len(tract[tract.amenity.isin(Amen_Fac)])
    row['Amenity-Waste Management']=len(tract[tract.amenity.isin(Amen_WM)])
    row['Amenity-Other']=len(tract[tract.amenity.isin(Amen_Other)])
    row['Barrier-Linear']=len(tract[tract.barrier.isin(bar_line)])
    row['Barrier-Highway Access Control']=len(tract[tract.barrier.isin(bar_AC)])
    row['Barrier-undefined']=len(tract[tract.barrier.isin(['yes'])])
    row['Barrier-Other']=len(tract[tract.barrier.isin(bar_other)])
    row['Building-Accomodation']=len(tract[tract.building.isin(build_Acc)])
    row['Building-Commercial']=len(tract[tract.building.isin(build_Com)])
    row['Building-Religious']=len(tract[tract.building.isin(build_Rel)])
    row['Building-Civic/amenity']=len(tract[tract.building.isin(build_CA)])
    row['Building-Agricultural']=len(tract[tract.building.isin(build_Agr)])
    row['Building-Sports']=len(tract[tract.building.isin(build_Sports)])
    row['Building-Storage']=len(tract[tract.building.isin(build_Stor)])
    row['Building-Cars']=len(tract[tract.building.isin(build_car)])
    row['Building-Power']=len(tract[tract.building.isin(build_pow)])
    row['Building-Other']=len(tract[tract.building.isin(build_other)])
    row['Building-Undefined']=len(tract[tract.building.isin(['yes'])])
    row['Craft']=len(tract[tract.craft.notnull()])
    row['Geological']=len(tract[tract.geological.notnull()])
    row['Historic']=len(tract[tract.historic.notnull()])
    row['Landuse-Developed']=len(tract[tract.landuse.isin(land_Dev)])
    row['Landuse-Rural']=len(tract[tract.landuse.isin(land_Rural)])
    row['Landuse-Waterbody']=len(tract[tract.landuse.isin(['basin','salt_pond'])])
    row['Landuse-Other']=len(tract[tract.landuse.isin(land_other)])
    row['Leisure']=len(tract[tract.leisure.notnull()])
    row['Man_Made']=len(tract[tract.man_made.notnull()])
    row['Military']=len(tract[tract.military.notnull()])
    row['Natural-Vegetation']=len(tract[tract.natural.isin(natural_veg)])
    row['Natural-Geology']=len(tract[tract.natural.isin(natural_geo)])
    row['Office']=len(tract[tract.office.notnull()])
    row['Place-Urban']=len(tract[tract.place.isin(place_urban)])
    row['Place-Rural']=len(tract[tract.place.isin(place_rural)])
    row['Shop-Food']=len(tract[tract.shop.isin(shop_food)])
    row['Shop-General']=len(tract[tract.shop.isin(shop_general)])
    row['Shop-Clothing']=len(tract[tract.shop.isin(shop_clothing)])
    row['Shop-Discount']=len(tract[tract.shop.isin(shop_discount)])
    row['Shop-Health']=len(tract[tract.shop.isin(shop_health)])
    row['Shop-D.I.Y.']=len(tract[tract.shop.isin(shop_diy)])
    row['Shop-Furniture']=len(tract[tract.shop.isin(shop_furniture)])
    row['Shop-Electronics']=len(tract[tract.shop.isin(shop_electronics)])
    row['Shop-Outdoors']=len(tract[tract.shop.isin(shop_outdoors)])
    row['Shop-Hobbies']=len(tract[tract.shop.isin(shop_hobby)])
    row['Shop-Gifts']=len(tract[tract.shop.isin(shop_gifts)])
    row['Shop-Other']=len(tract[tract.shop.isin(shop_other)])
    row['Sport']=len(tract[tract.sport.notnull()])
    row['Tourism']=len(tract[tract.tourism.notnull()])
    row['Motorway']=0
    row['Trunk']=0
    row['Primary']=0
    row['Secondary']=0
    row['Tertiary']=0
    row['Unclassified Highway']=0
    row['Residential Highway']=0
    row['Special Roads']=0
    row['Unknown Roads']=0
    row['Paths']=0
    row['Waterway-Natural']=0
    row['Waterway-Man Made']=0
    row['Waterway-Facilities']=0
    row['Waterway-Barriers']=0
    row['Waterway-Other']=0
    row['Aerialway']=0
    row['Railway_Tracks']=0
    row['Area']=int(tract.Area.unique()[0])
    row['GEOID']=tract.GEOID.unique()[0]
    row['Tract']=tract.Tract.unique()[0]
    row = {k:[v] for k,v in row.items()}
    return row
```

RUN MAX CARDINALITY MATCHING ALGORITHM ON graph.txt

Read through output best matching

```python
c=0
best_edges=[]
with open("C:\\Users\\nairk\\census-by-precincts\\info\\"+folder+"\\"+city+"\\"+inits+"_Matching.txt", 'r',encoding='UTF-16 LE') as file:
    for line in file:
        if c==2:
            edge=line.replace('\n','')
            nodes=edge.split()
            best_edges.append([list(dict.keys())[list(dict.values()).index(int(nodes[0]))],list(dict.keys())[list(dict.values()).index(int(nodes[1]))]])
        else:
            if c==0:
                print(line)
            c+=1
```

Check it is a matching

```python
v=[]
test=[]
for i in best_edges:
    for j in i:
        test.append(j)
        if j not in v:
            v.append(j)
        else:
            print("error")
```

Combine two paired tract tuples into a single tuple

```python
for e in best_edges:
    tract1=pd.read_csv(dir+"\\Tract_"+e[0]+".csv",dtype={'GEOID': object})
    tract2=pd.read_csv(dir+"\\Tract_"+e[1]+".csv",dtype={'GEOID': object})
    area=tract1["Area"][0]+tract2["Area"][0]
    area_ratio=tract1["Area"][0]/area
    pop1=int(pops[pops["GEO_ID"]==tract1["GEOID"][0]]["P1_001N"].iloc[0])
    tot_pop=pop1+int(pops[pops["GEO_ID"]==tract2["GEOID"][0]]["P1_001N"].iloc[0])
    name=pops[pops["GEO_ID"]==tract1["GEOID"][0]]["NAME"].iloc[0] +'$'+pops[pops["GEO_ID"]==tract2["GEOID"][0]]["NAME"].iloc[0]
    urban= int(housing2[housing2["GEO_ID"]==tract1["GEOID"][0]]["H2_002N"].iloc[0])+int(housing2[housing2["GEO_ID"]==tract2["GEOID"][0]]["H2_002N"].iloc[0])
    rural = int(housing2[housing2["GEO_ID"]==tract1["GEOID"][0]]["H2_003N"].iloc[0])+int(housing2[housing2["GEO_ID"]==tract2["GEOID"][0]]["H2_003N"].iloc[0])
    occupied=int(housing[housing["GEO_ID"]==tract1["GEOID"][0]]["H1_002N"].iloc[0])+int(housing[housing["GEO_ID"]==tract2["GEOID"][0]]["H1_002N"].iloc[0])
    vacant=int(housing[housing["GEO_ID"]==tract1["GEOID"][0]]["H1_003N"].iloc[0])+int(housing[housing["GEO_ID"]==tract2["GEOID"][0]]["H1_003N"].iloc[0])
    h1=int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_003N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_011N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_003N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_011N"].iloc[0])
    h2=int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_004N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_012N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_004N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_012N"].iloc[0])
    h3=int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_005N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_013N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_005N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_013N"].iloc[0])
    h4=int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_006N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_014N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_006N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_014N"].iloc[0])
    h5=int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_007N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_015N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_007N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_015N"].iloc[0])
    h6=int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_008N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_016N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_008N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_016N"].iloc[0])
    h7=int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_009N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract1["GEOID"][0]]["H12_017N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_009N"].iloc[0])+int(housing12[housing12["GEO_ID"]==tract2["GEOID"][0]]["H12_017N"].iloc[0])
    tract1=tract1.drop(columns=["Area","GEOID","Tract"])
    tract2=tract2.drop(columns=["Area","GEOID","Tract"])
    tract1=tract1.add_suffix("_1")
    tract2=tract2.add_suffix("_2")
    tract=pd.concat([tract1,tract2],axis=1)
    tract["Total_Area"]=area
    tract["Target_Area_Ratio"]=area_ratio
    tract["Total_Pop"]=tot_pop
    tract["Combined_Name"]=name
    tract["Urban"] = urban
    tract["Rural"] = rural
    tract["Occupied_Housing"]=occupied
    tract["Vacant_Housing"]=vacant
    tract["One_Housing"] = h1
    tract["Two_Housing"] = h2
    tract["Three_Housing"] = h3
    tract["Four_Housing"] = h4
    tract["Five_Housing"] = h5
    tract["Six_Housing"] = h6
    tract["Seven_Plus_Housing"] = h7
    if tot_pop==0:
        tract["Target"]=0.0
    else:
        tract["Target"]=pop1/tot_pop
    tract.to_csv(out_dir+"\\"+name+".csv",index=False)
```
